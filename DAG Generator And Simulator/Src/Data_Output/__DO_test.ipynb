```python
import graphviz                     # 创建树对象
import numpy as np
import matplotlib.pyplot as plt     # plt 用于显示图片
import matplotlib.image as mpimg    # mpimg 用于读取图片
from matplotlib.font_manager import FontProperties
```

```python
mygraph = graphviz.Digraph(node_attr={'shape': 'box'}, edge_attr={'labeldistance': "10.5"}, format="png")
```

```python
for node_id, node_x in enumerate(["Has feathers?", "Can fly?", "Has fins?", "Hawk", "Penguin", "Dolphin", "Bear"]):
    mygraph.node(f'{node_id}', node_x)      # 构建节点
```

```python
for edge_x in ([(0, 1, True), (0, 2, False), (1, 3, True), (1, 4, False), (2, 5, True), (2, 6, False)]):
    mygraph.edge(f'{edge_x[0]}', f'{edge_x[1]}', label=f'{edge_x[2]}')     # 构建边
```

```python
mygraph.render("./pic/decisionTree")  # 图形显示
plt.cla()       # Clear axis
plt.clf()       # (Clear figure)清除当前图形及其所有轴，但保持窗口打开，以便可以将其重新用于其他绘图。
plt.close()     # (Close a figure window)关闭一个窗口，如果没有另外指定，该窗口将是当前窗口
plt.figure(num=None, figsize= (12, 12), dpi=80, facecolor='w', edgecolor='k')

plt.title("DAG_Graph_Sample", fontsize=12, color="black", weight="light", ha='left', x=0)
plt.xlabel("time-axis", fontdict={'family': 'Times New Roman', 'size': 16})
# plt.xticks(np.arange(10),np.arange(10), fontproperties='Times New Roman', size=16, rotation=30)
# plt.xticks([-1, 0, 1], ['a','b','c'], fontproperties='Times New Roman', size=16, rotation=30)
# plt.xlim(xmin=-2, xmax=2)

plt.ylabel("core-list", fontdict={'family': 'Times New Roman', 'size': 16})
# plt.yticks([1.5, 2.5, 3.5], ['a', 'b', 'c'], fontproperties='Times New Roman', size=16)  # 设置y刻度:用文字来显示刻度
plt.axvline(x=0, ls="-", c="red")  # 添加垂直直线
# plt.xlim(xmin=0, xmax=2260000)

# 右上标记 legend(xxx)
did_list = ['dag1','dag2']
color_list = ['#a6cee3', '#1f78b4']
obj_list = [plt.scatter(0, 0, marker="s", color=color_list[k]) for k, v in enumerate(did_list)]
plt.legend(obj_list, did_list, loc='upper right', prop={'family': 'Times New Roman', 'size': 16})
ax = plt.gca()                          # 获取图形坐标轴

# font = FontProperties(size=16)
# plt.legend(temp_obj_list, Dag_ID_List, loc='upper right', prop=font)
# ax.ticklabel_format(style='plain')

# ax.set_axis_off()                     # 去掉坐标
decisionTree = mpimg.imread('./pic/decisionTree.png') # 读取和代码处于同一目录下的 lena.png
ax.imshow(decisionTree)
plt.show()
```

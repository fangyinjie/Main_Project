## 3 DAG调度仿真

```python
# DAG初始化
TTL = 1130000
core_num = 5
Total_Time = 20 * TTL
for dag_id, dag_x in enumerate(All_DAG_list):
    DFA.dag_data_initial(dag_x, DAGType=int(dag_id), DAG_id=int(dag_id), Period=Total_Time/2, Critic=1)
    DPC.Priority_Config('SELF', dag_x)
```

```python
# DAG执行
Dispatcher = SS.Dispatcher_Workspace([copy.deepcopy(All_DAG_list), {'Core_Num': core_num, 'Total_Time': Total_Time, 'Enqueue_rank': False, 'Priority_rank': True, 'Preempt_type': False, 'Dynamic': False}])
Dispatcher.run()

SELF_h = Dispatcher.Core_Data_List
makespan = Core.ret_makespan(SELF_h)
workload_rade = 100 * sum([DFA.get_dag_volume(dag_x) for dag_x in All_DAG_list]) / sum([core_x.get_core_last_time() for core_id, core_x in SELF_h.items()])

print(f'makespan:{makespan}')
print(f'workload_rade:{workload_rade}')
SRS.show_core_data_list({'dag': SELF_h}, 'Show', '')
```
